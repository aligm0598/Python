```python
import pandas as pd
```

```python
delays_df = pd.read_csv('Data/flight_delays.csv') 
```

```python
import matplotlib.pyplot as plt
```

```python
delays_df.plot(
               kind='scatter',
               x='DISTANCE',
               y='ARRIVAL_DELAY',
               color='blue',
               alpha=0.3,
               title='Correlation of arrival and distance'
              )
plt.show()
```

```python
delays_df.plot(
               kind='scatter',
               x='DEPARTURE_DELAY',
               y='ARRIVAL_DELAY',
               color='blue',
               alpha=0.3,
               title='Correlation of arrival and departure delay'
              )
plt.show()
```
